```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_algorithm(data):
    fig, ax = plt.subplots()
    x = data.Generation
    ymax = data.Max
    ymin = data.Min
    ymean = data.Mean
    ax.plot(x, ymax, label = "Maximum score")
    ax.plot(x, ymin, label = "Minimum score")
    ax.plot(x, ymean, label = "Average score")
    ax.legend(fontsize = "medium")
    ax.set_xlabel("Generation")
    ax.set_ylabel("Score")

    plt.show()
```

```python
katsuura = pd.read_csv('plotKatsuura.csv', delimiter=';')
plot_algorithm(katsuura)
```

```python
schaffers = pd.read_csv('plotSchaffers.csv', delimiter=';')
plot_algorithm(schaffers)
```
